```python
import matplotlib.pyplot as plt
import seaborn as sns

features = ['Sociability', 'Solo_preference', 'Talkative', 'Grouping', 'Partying',
            'Listening', 'Speaking', 'Curiosity', 'Routine', 'Thrill', 
            'Friendly', 'Adventure', 'Stress']

plt.figure(figsize=(14, 8))
sns.boxplot(data=df[features], orient='h')
plt.title("Outlier Detection via Boxplot")
plt.tight_layout()
plt.show()
```

```python
Q1 = df[features].quantile(0.25)
Q3 = df[features].quantile(0.75)
IQR = Q3 - Q1

outlier_mask = ((df[features] < (Q1 - 1.5 * IQR)) | (df[features] > (Q3 + 1.5 * IQR))).any(axis=1)
print(f"Total outlier rows (IQR method): {outlier_mask.sum()}")

# Show a few
df[outlier_mask].head()
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
df["Personality"] = le.fit_transform(df["Personality"])
```

```python
cols = [
    'Sociability', 'Solo_preference', 'Talkative', 'Grouping', 'Partying',
    'Listening', 'Speaking', 'Curiosity', 'Routine', 'Thrill',
    'Friendly', 'Adventure', 'Stress'
]

# Convert to int (round first to avoid decimals)
df[cols] = df[cols].astype(int)

print(df[cols].dtypes.head())
```

```python
from sklearn.preprocessing import StandardScaler

features = [
    'Sociability', 'Solo_preference', 'Talkative', 'Grouping', 'Partying',
    'Listening', 'Speaking', 'Curiosity', 'Routine', 'Thrill',
    'Friendly', 'Adventure', 'Stress'
]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Define features and target
X = df[['Sociability', 'Solo_preference', 'Talkative', 'Grouping', 'Partying',
        'Listening', 'Speaking', 'Curiosity', 'Routine','Thrill','Friendly','Adventure','Stress']]
y = df['Personality']  # already encoded (0,1,2)

# Split dataset
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print(X_train.shape, X_val.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predictions
y_pred = model.predict(X_val_scaled)

# Metrics
print("Accuracy:", accuracy_score(y_val, y_pred))
print("\nClassification Report:\n", classification_report(y_val, y_pred))

# Confusion matrix
cm = confusion_matrix(y_val, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
import pandas as pd
import numpy as np

importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': np.mean(np.abs(model.coef_), axis=0)
}).sort_values('Coefficient', ascending=False)

print(importance)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
print("RandomForest Accuracy:", accuracy_score(y_val, rf.predict(X_val)))
```
